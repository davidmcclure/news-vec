# A-vs-B accuracies, baseline models

```python
import pandas as pd

from news_vec.utils import read_json_gz_lines
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import altair as alt
import seaborn as sns

mpl.style.use('seaborn-muted')
sns.set(style="whitegrid")
```

```python
df = pd.DataFrame(read_json_gz_lines('../../data/clf-ab-acc.json/'))
```

```python
df['label'] = df[['d1', 'd2']].apply(lambda x: f'{x[0]} - {x[1]}', axis=1)
```

```python
order = df.groupby('label').mean().acc_lr.sort_values(ascending=False).index
plt.figure(figsize=(5,30))
sns.stripplot(x='acc_lr', y='label', data=df, order=order, s=2)
```

```python
order = df.groupby('label').mean().acc_svc.sort_values(ascending=False).index
plt.figure(figsize=(5,30))
sns.stripplot(x='acc_svc', y='label', data=df, order=order, s=2)
```
