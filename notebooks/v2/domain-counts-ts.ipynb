```python
import pandas as pd

from IPython.display import display, Markdown
from datetime import date
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import altair as alt
import seaborn as sns

mpl.style.use('seaborn-muted')
sns.set(style="whitegrid")
```

```python
pd.options.display.max_rows = 200
```

```python
def load_ts_df(path):
    df = pd.read_json(path, lines=True)
    df['date'] = df[['year', 'month', 'day']].apply(lambda x: date(*x), axis=1)
    return df.sort_values('date')
```

```python
def plot_ts(df, w=15, h=3):
    plt.figure(figsize=(w,h))
    plt.bar(list(df.date), df['count'], linewidth=0, width=1)
    plt.show()
```

```python
links = load_ts_df('data/domain-link-ts-counts.json/part-00000-fd8939b8-5608-4582-b944-ff69225f8912-c000.json')
```

```python
actors = load_ts_df('data/domain-actor-ts-counts.json/part-00000-719a8558-d4f3-4b20-bfce-c65a70f3a9a0-c000.json')
```

```python
articles = load_ts_df('data/domain-article-ts-counts.json/part-00000-f4b25eba-b117-4525-9b2c-e0181fe8f520-c000.json')
```

```python
link_counts = pd.read_json('data/domain-link-counts.json/part-00000-d73512e3-ea00-4314-a7b9-10ddcd09b8b2-c000.json', lines=True)
```

```python
for d in link_counts.domain:
    display(Markdown(f'# {d}'))
    print('Links')
    plot_ts(links[links.domain==d])
    print('Actors')
    plot_ts(actors[actors.domain==d])
    print('Articles')
    plot_ts(articles[articles.domain==d])
```
