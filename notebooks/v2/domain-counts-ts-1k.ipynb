```python
import pandas as pd

from scipy import stats
from IPython.display import display, Markdown
from datetime import date, datetime as dt
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import altair as alt
import seaborn as sns

mpl.style.use('seaborn-muted')
sns.set(style="whitegrid")
```

```python
pd.options.display.max_rows = 200
```

```python
def load_ts_df(path):
    df = pd.read_json(path, lines=True)
    df['date'] = df[['year', 'month', 'day']].apply(lambda x: date(*x), axis=1)
    return df.sort_values('date')
```

```python
def plot_ts(df, y='count', w=15, h=3):
    plt.figure(figsize=(w,h))
    plt.bar(list(df.date), df[y], linewidth=0, width=1)
    plt.show()
```

```python
links = load_ts_df('data/domain-link-ts-counts-1k.json/part-00000-57f9be8e-1953-4f0b-85ec-aee10f36abdf-c000.json')
```

```python
articles = load_ts_df('data/domain-article-ts-counts-1k.json/part-00000-99ef162e-96dc-4de2-8773-d0e6806bc91c-c000.json')
```

```python
link_counts = pd.read_json('data/domain-link-counts-1k.json/part-00000-e8ec26f3-6cdd-4bec-8639-f49270871a8b-c000.json', lines=True)
```

```python
for d in link_counts.domain:
    display(Markdown(f'# {d}'))
    print('Links')
    plot_ts(links[links.domain==d])
    print('Articles')
    plot_ts(articles[articles.domain==d])
```
