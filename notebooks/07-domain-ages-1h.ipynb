```python
import pandas as pd
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.style.use('seaborn-muted')
sns.set(style="whitegrid")

%matplotlib inline
```

```python
ages = pd.read_json('../data/domain-ages-1h.json/part-00000-65898407-51f2-450c-a794-27aad87fa8ee-c000.json', lines=True)
```

```python
ages.head(10)
```

```python
ages.sort_values('bin', inplace=True)
```

```python
ages.domain.unique()
```

# 1 hour

```python
df = ages[ages.bin>10]

plt.figure(figsize=(15,10))
# plt.yscale('log')

# for domain in ages.domain.unique()[:1]:
#     print(domain)
for d in ('nytimes.com', 'cnn.com', 'reuters.com', 'huffingtonpost.com', 'buzzfeed.com', 'rt.com', 'politico.com'):
    ddf = df[df.domain==d]
    counts = ddf['count']
    plt.plot(ddf.bin, counts, linewidth=1, alpha=0.5, label=d)
    
plt.legend()
```

```python
plt.figure(figsize=(15,10))

for d in ('nytimes.com', 'cnn.com', 'reuters.com', 'politico.com'):
    ddf = ages[(ages.domain==d) & (ages.bin>10)]
    counts = ddf['count'] / ages[ages.domain==d]['count'].sum()
    plt.plot(ddf.bin, counts, linewidth=1, alpha=0.8, label=d)
    
plt.legend()
```
