```python
import pandas as pd
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.style.use('seaborn-muted')
sns.set(style="whitegrid")

%matplotlib inline
```

```python
ages = pd.read_json('../data/ages-1h.json/part-00000-7a779745-1593-4f44-9199-3a14db8b8143-c000.json', lines=True)
```

```python
ages.head(10)
```

```python
ages.sort_values('bin', inplace=True)
```

# 1 hour

```python
df = ages[ages.bin>0]
plt.figure(figsize=(15,10))
_ = plt.plot(df.bin, df['count'], linewidth=1)
```

```python
df = ages[ages.bin>0]
plt.figure(figsize=(15,10))
plt.yscale('log')
_ = plt.plot(df.bin, df['count'], linewidth=1)
```

# 10 minutes

```python
df = ages[(ages.bin>0) & (ages.bin<600)]
plt.figure(figsize=(15,10))
plt.yscale('log')
_ = plt.plot(df.bin, df['count'], linewidth=2)
```
