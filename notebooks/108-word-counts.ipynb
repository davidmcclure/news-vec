```python
import pandas as pd

from news_vec.corpus import HeadlineDataset
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import altair as alt
import seaborn as sns

mpl.style.use('seaborn-muted')
sns.set(style="whitegrid")
```

```python
ds = HeadlineDataset.load('../data/ava.p')
```

```python
df = pd.DataFrame([r for r, _ in ds.train])
```

```python
len(df)
```

```python
df['wc'] = df.clf_tokens.apply(lambda ts: len(ts))
```

```python
df['wc_rank'] = df.wc.rank(pct=True)
```

```python
pdf = df
order = pdf.groupby('domain').mean().wc.sort_values().index
plt.figure(figsize=(6,10))
plt.xlim(0, 30)
sns.boxplot(x='wc', y='domain', data=pdf, order=order)
```
