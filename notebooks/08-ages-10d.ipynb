```python
import pandas as pd
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.style.use('seaborn-muted')
sns.set(style="whitegrid")

%matplotlib inline
```

```python
ages = pd.read_json('../data/ages-10d.json/part-00000-33cd4fbc-b1c5-43a4-b796-5595c2fcf80e-c000.json', lines=True)
```

```python
ages.sort_values('bin', inplace=True)
```

```python
ages.head(10)
```

# 10 days

```python
plt.figure(figsize=(15,10))
_ = plt.plot(ages.bin, ages['count'], linewidth=1)
```

```python
ticks = range(0, ages.bin.max(), 60*60*24)

plt.figure(figsize=(15,10))
plt.xticks(ticks)
plt.yscale('log')
_ = plt.plot(ages.bin, ages['count'], linewidth=1)
```
