```python
import gzip
import ujson

from itertools import islice
from glob import glob

from news_vec.headlines import parse_headline
```

```python
def read_json_gz_lines(root):
    """Read JSON corpus.

    Yields: Line
    """
    for path in glob('%s/*.gz' % root):
        with gzip.open(path) as fh:
            for line in fh:
                yield ujson.loads(line)
```

```python
lines = list(islice(read_json_gz_lines('../data/b13-texts.json/'), 1000))
```

```python
for line in lines[:200]:
    doc = parse_headline(line['title'])
    print(line['label'])
    print(line['title'])
    print(' '.join(doc._.clf_tokens), '\n')
```
