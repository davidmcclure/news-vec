```python
import pandas as pd

from collections import OrderedDict
from sqlalchemy.sql import func

from news_vec.link_db import Link, session
```

```python
pd.options.display.max_rows = 200
```

```python
count = func.count(Link.id)

query = (session
    .query(Link.domain, count)
    .group_by(Link.domain)
    .order_by(count.desc()))

rows = query.all()

link_counts = pd.DataFrame(rows, columns=('domain', 'count'))
```

```python
link_counts
```

```python
count = func.count(Link.article_id.distinct())

query = (session
    .query(Link.domain, count)
    .group_by(Link.domain)
    .order_by(count.desc()))

rows = query.all()

article_counts = pd.DataFrame(rows, columns=('domain', 'count'))
```
