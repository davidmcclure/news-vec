```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import altair as alt
import seaborn as sns

mpl.style.use('seaborn-muted')
sns.set(style="whitegrid")

%matplotlib inline
```

```python
df, embeds = read_preds('data/ava-preds/')
```

```python
df['p_true'] = df.apply(lambda r: r[f'p_{r.domain}'], axis=1)
```

```python
df.groupby('domain').size()
```

```python
sns.distplot(df[df.split=='train'].p_true, bins=50)
plt.axvline(0.5)
```

```python
for d in df.domain.unique():
    print(d)
    sns.distplot(df[df.domain==d].p_true, bins=50)
    plt.axvline(0.5)
    plt.show()
```
