```python
import pandas as pd

from tqdm import tqdm

from news_vec.utils import read_json_gz_lines
```

```python
df = pd.DataFrame(list(tqdm(read_json_gz_lines('../data/clf-articles.json/'))))
```

```python
df['clf_token_len_ratio'] = df.apply(lambda r: len(r.clf_tokens)/ len(r.tokens), 1)
```

```python
df.groupby('domain').mean().sort_values('clf_token_len_ratio')
```
