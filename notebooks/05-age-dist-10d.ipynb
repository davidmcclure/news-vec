```python
import pandas as pd
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.style.use('seaborn-muted')
sns.set(style="whitegrid")

%matplotlib inline
```

```python
counts = pd.read_csv('../data/age-10d.csv/part-00000-4a399ca2-85ea-4ed3-a6a9-8129379bd823-c000.csv')
```

```python
counts.head(10)
```

```python
plt.figure(figsize=(10,5))
top = counts
_ = plt.hist(top.age, weights=top['count'], bins=200, linewidth=0)
```

```python
plt.figure(figsize=(10,5))
plt.yscale('log')
top = counts
_ = plt.hist(top.age, weights=top['count'], bins=200, linewidth=0)
```

```python
plt.figure(figsize=(10,5))
top = counts[counts.age>0].head(60*10)
_ = plt.hist(top.age, weights=top['count'], bins=60*10, linewidth=0, width=1)
```

```python
plt.figure(figsize=(10,5))
plt.yscale('log')
top = counts[counts.age>0].head(60*10)
_ = plt.hist(top.age, weights=top['count'], bins=60*10, linewidth=0, width=1)
```

```python
plt.figure(figsize=(20,5))
plt.yscale('log')
plt.xticks(range(0, 60*5, 60))
top = counts[counts.age>0].head(60*5)
_ = plt.hist(top.age, weights=top['count'], bins=60*5, linewidth=0, width=1)
```
