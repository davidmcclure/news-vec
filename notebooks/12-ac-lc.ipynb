```python
import pandas as pd
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

import altair as alt
import seaborn as sns

mpl.style.use('seaborn-muted')
sns.set(style="whitegrid")

%matplotlib inline
```

```python
pd.options.display.max_rows = 100
```

```python
lc = pd.read_json('../data/lc.json/part-00000-8983a7ec-c881-4dea-9e57-f187cf438ee1-c000.json', lines=True)
```

```python
ac = pd.read_json('../data/ac.json/part-00000-cbe5246b-ad53-4782-9ab8-68e0e5400505-c000.json', lines=True)
```

```python
lc.sort_values('count', ascending=False)
```

```python
ac.sort_values('count', ascending=False)
```

```python
counts = lc.rename(columns={'count': 'lc_count'})
```

```python
counts = counts.merge(ac.rename(columns={'count': 'ac_count'}))
```

```python
counts['ratio'] = counts.ac_count / counts.lc_count
```

```python
counts.sort_values('ratio')
```

```python
order = counts.sort_values('ratio', ascending=True).domain
plt.figure(figsize=(8,30))
plt.xlim(0, 1)
sns.barplot(x=counts.ratio, y=counts.domain, order=order)
```

```python
chart = (alt
    .Chart(counts)
    .mark_point()
    .encode(
        x=alt.X('lc_count', scale=alt.Scale(type='log')),
        y='ratio',
        tooltip='domain',
    ))
```
